```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url="https://in.finance.yahoo.com/quote/BANKBARODA.BO/history/"
data_link=requests.get(url)
```

```python
data_link.text
```

```python
final_html = bs(data_link.text, "html.parser")
```

```python
data=[]
```

```python
for tabel in final_html.find("table",{"class":"W(100%) M(0)"}):
    for tbody in tabel:
        for tr in tbody:
            data.append(tr.get_text())
```

```python
data
```

```python
seperate_rows=[]
```

```python
rows=len(data)/7
for x in range(round(rows)):
    new_temp=[]
    for i in range(7):
        new_temp.append(data[0])
        data.pop(0)
    seperate_rows.append(new_temp)
```

```python
seperate_rows
```

```python
df1=pd.DataFrame(seperate_rows)
```

```python
df1.columns=['Date','Open','High','Low','Close*','Adj. close**','Volume']
```

```python
df1
```

```python
data=df1.iloc[1:]
```

```python
data
```

```python
correct=[]
for i in data['Volume']:
    a=i.replace(",","")
    correct.append(a)
```

```python
correct
```

```python
correct=correct.pop(27)
```

```python
data['Date']=data['Date'].pop(27)
data['Open']=data['Open'].pop(27)
data['High']=data['High'].pop(27)
data['Low']=data['Low'].pop(27)
data['Close*']=data['Close*'].pop(27)
data['Adj. close**']=data['Adj. close**'].pop(27)
```

```python
data['Volume_int']=correct
```

```python
data=data.drop(columns=['Volume'],axis=1)
```

```python
data
```

```python
data.info()
```

```python
data['Volume_int']=data['Volume_int'].astype(int)
data['Open']=data['Open'].astype(float)
data['High']=data['High'].astype(float)
data['Low']=data['Low'].astype(float)
data['Close*']=data['Close*'].astype(float)
data['Adj. close**']=data['Adj. close**'].astype(float)
```
